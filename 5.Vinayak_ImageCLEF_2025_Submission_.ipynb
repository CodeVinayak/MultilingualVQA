```python
import os
import json
import re
import gc
import torch
import time # For basic timing
from PIL import Image
from datasets import load_dataset, load_from_disk
from transformers import AutoProcessor
from unsloth import FastVisionModel
from peft import PeftModel
import traceback # For detailed error logging

# --- Configuration ---
DATASET_NAME = "MBZUAI/EXAMS-V"
DATASET_SPLIT = "test"
dataset_load_path = f"./{DATASET_NAME.split('/')[-1]}_{DATASET_SPLIT}" # Path where dataset is saved/loaded from

OUTPUT_JSONL_FILENAME = "run_results_unstructured.jsonl" # Use JSON Lines format for incremental saving
FINAL_JSON_FILENAME = "run.json" # Final output file in the required format

# Image Resizing Configuration
MAX_PIXELS = 1280 * 28 * 28 # Max pixels allowed (adjust if needed, lower values save memory)
# MAX_PIXELS = 1024 * 1024 # Example: Lower limit if needed

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {DEVICE}")

# Model Configuration (Keep your existing config)
BASE_MODEL_ID = "unsloth/Qwen2.5-VL-7B-Instruct-unsloth-bnb-4bit"
ADAPTER_PATH = "/content/vinayak_model" # Make sure this path is correct
PROCESSOR_ID = "Qwen/Qwen2.5-VL-7B-Instruct"

# --- VLM Prompt ---
VLM_PROMPT = (
    "Analyze the image of a multiple-choice question. Identify the question, "
    "all answer options (even if there are more than four), and any relevant "
    "visuals like graphs or tables. Choose the correct answer based only on the image. "
    "Reply with just the letter of the correct option, no explanation."
)

# --- Helper Functions ---

def resize_image(img: Image.Image, max_pixels: int) -> Image.Image:
    """Resizes a PIL image to not exceed max_pixels while maintaining aspect ratio."""
    try:
        width, height = img.size
        current_pixels = width * height

        if current_pixels <= max_pixels:
            # print(f"Image already within pixel limit ({width}x{height}).")
            return img

        scale_factor = (max_pixels / current_pixels) ** 0.5
        new_width = max(1, int(width * scale_factor)) # Ensure width >= 1
        new_height = max(1, int(height * scale_factor)) # Ensure height >= 1

        print(f"Resizing image from {width}x{height} to {new_width}x{new_height}...")
        # LANCZOS is high quality but can be slower; ANTIALIAS is often faster and good
        # img_resized = img.resize((new_width, new_height), Image.Resampling.LANCZOS)
        img_resized = img.resize((new_width, new_height), Image.Resampling.NEAREST) #faster

        return img_resized
    except Exception as e:
        print(f"Error during image resizing: {e}")
        return img # Return original image on error

# MODIFIED FUNCTION: extract_predicted_answer
def extract_predicted_answer(model_output_text):
    """
    Attempts to extract a single uppercase letter (A-E) from model output.
    Handles common variations like lowercase or surrounding text.
    If no standard A-E answer can be reliably extracted, returns the original
    (trimmed) model output text.
    """
    text = model_output_text.strip()

    # --- First, check for exact single uppercase letter ---
    if text in ["A", "B", "C", "D", "E"]:
        return text

    # --- NEW: Check for exact single lowercase letter ---
    if text.lower() in ["a", "b", "c", "d", "e"] and len(text) == 1:
        print(f"Info: Converted lowercase single letter '{text}' to uppercase.")
        return text.upper()

    # --- Then, try common patterns (case-insensitive where appropriate) ---
    # Pattern: Quoted letter 'A', "B", `C`, 'a', "b" etc.
    match = re.search(r'[`\'"]([ABCDE])[`\'"]', text, re.IGNORECASE)
    if match:
        print(f"Info: Extracted letter '{match.group(1).upper()}' from quoted pattern in: '{text}'.")
        return match.group(1).upper()

    # Pattern: "answer is A", "option: B", "choice c" etc. (case-insensitive)
    match = re.search(r'(?:answer|option|choice) is\s*:?\s*([ABCDE])', text, re.IGNORECASE)
    if match:
        print(f"Info: Extracted letter '{match.group(1).upper()}' from pattern like 'answer is X' in: '{text}'.")
        return match.group(1).upper()

    # --- Look for standalone letters (less reliable, check uppercase first) ---
    # Find all standalone uppercase letters A-E
    matches_upper = re.findall(r'\b([ABCDE])\b', text)
    if matches_upper:
        # Prefer the last match found.
        print(f"Warning: Extracted letter '{matches_upper[-1]}' based on standalone UPPERCASE occurrence in: '{text}'. Using it.")
        return matches_upper[-1] # Already uppercase

    # Find all standalone lowercase letters a-e (only if no uppercase found)
    matches_lower = re.findall(r'\b([abcde])\b', text)
    if matches_lower:
        # Prefer the last match found and convert to uppercase.
        print(f"Warning: Extracted letter '{matches_lower[-1].upper()}' based on standalone LOWERCASE occurrence in: '{text}'. Using it.")
        return matches_lower[-1].upper()

    # --- If all extraction attempts fail ---
    print(f"Warning: Could not reliably extract a standard answer (A-E) from: '{text}'. Returning raw output.")
    # MODIFICATION: Return the original trimmed text instead of None
    return text

def load_processed_ids(filename: str) -> set:
    """Loads sample_ids from a JSON Lines file into a set for quick checking."""
    processed_ids = set()
    if not os.path.exists(filename):
        return processed_ids
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            for line in f:
                try:
                    data = json.loads(line)
                    # Use 'id' because that's the key we write to the file
                    if 'id' in data:
                        processed_ids.add(data['id'])
                except json.JSONDecodeError:
                    print(f"Warning: Skipping malformed line in {filename}: {line.strip()}")
    except Exception as e:
        print(f"Error loading processed IDs from {filename}: {e}")
    print(f"Loaded {len(processed_ids)} already processed IDs from {filename}.")
    return processed_ids

def append_result_to_file(filename: str, result_dict: dict):
    """Appends a single result dictionary to a JSON Lines file."""
    try:
        with open(filename, 'a', encoding='utf-8') as f:
            json.dump(result_dict, f, ensure_ascii=False)
            f.write('\n')
    except Exception as e:
        print(f"Error appending result to {filename}: {e}")

def clear_memory():
    """Explicitly runs garbage collection and clears GPU cache."""
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
    # print("Memory cleared.") # Optional: uncomment for verbose output

# --- Main Processing ---

print("--- Starting Script ---")

# 1. Load Model, Tokenizer, and Processor
print("Loading model, tokenizer, and processor...")
try:
    model, tokenizer = FastVisionModel.from_pretrained(
        BASE_MODEL_ID,
        load_in_4bit=True,
        use_gradient_checkpointing="unsloth",
        # Consider adding device_map if you have multiple GPUs or specific needs
        # device_map="auto"
    )
    print(f"Loading PEFT adapter from: {ADAPTER_PATH}")
    model = PeftModel.from_pretrained(model, ADAPTER_PATH)
    print("PEFT adapter loaded successfully.")
    print(f"Loading processor: {PROCESSOR_ID}")
    processor = AutoProcessor.from_pretrained(PROCESSOR_ID, trust_remote_code=True)
    print("Model components loaded successfully.")
    model.eval() # Set model to evaluation mode immediately
    print(f"Model is in evaluation mode. Device: {DEVICE}")
except Exception as e:
    print(f"FATAL: Error loading model/tokenizer/processor: {e}")
    print(traceback.format_exc())
    exit()

# 2. Load Dataset
print(f"\nLoading dataset '{dataset_load_path}' (split: {DATASET_SPLIT})...")
try:
    if os.path.isdir(dataset_load_path):
         test_dataset = load_from_disk(dataset_load_path)
    else:
         test_dataset = load_dataset(DATASET_NAME, split=DATASET_SPLIT)
         test_dataset.save_to_disk(dataset_load_path) # Save for future runs
    print(f"Dataset loaded. Number of examples: {len(test_dataset)}")
except Exception as e:
    print(f"FATAL: Error loading dataset: {e}")
    print(traceback.format_exc())
    exit()

# 3. Load IDs of already processed examples for resumability
processed_ids = load_processed_ids(OUTPUT_JSONL_FILENAME)
total_examples = len(test_dataset)
processed_count_this_run = 0
skipped_count = 0

print(f"\nStarting prediction loop for {total_examples} examples...")
print(f"Output will be incrementally saved to: {OUTPUT_JSONL_FILENAME}")

# 4. Process Dataset Incrementally
start_time_loop = time.time()

for i, example in enumerate(test_dataset):
    start_time_item = time.time()
    item_id = example.get("sample_id") # Use sample_id from dataset

    # --- Resumability Check ---
    if item_id in processed_ids:
        # print(f"--- Skipping example {i+1}/{total_examples} (ID: {item_id}) - Already processed ---")
        skipped_count += 1
        continue # Skip to the next item

    print(f"\n--- Processing example {i+1}/{total_examples} (ID: {item_id}) ---")

    # --- Get Data & Validate ---
    language = example.get("language", "Unknown") # Default if missing
    image_data = example.get("image")

    if not item_id:
        print(f"Warning: Skipping example {i} due to missing 'sample_id'.")
        clear_memory() # Clean memory even if skipped early
        continue
    if image_data is None:
        print(f"Warning: Skipping example {i} (ID: {item_id}) due to missing 'image'.")
        clear_memory()
        continue

    # --- Image Loading and Preprocessing ---
    try:
        # Ensure image is PIL object
        if not isinstance(image_data, Image.Image):
            if isinstance(image_data, str) and os.path.exists(image_data):
                image_pil = Image.open(image_data)
            else:
                 # Try loading from bytes if available (common in HF datasets)
                 try:
                     from io import BytesIO
                     image_pil = Image.open(BytesIO(image_data['bytes']))
                 except Exception:
                    raise ValueError("Image data is not a PIL Image, valid path, or recognized bytes format.")
        else:
            image_pil = image_data

        # Convert to RGB
        image_pil = image_pil.convert("RGB")

        # Resize image *before* passing to processor/model
        image_resized = resize_image(image_pil, MAX_PIXELS)
        print(f"Image Size after potential resize: {image_resized.size}")

    except Exception as img_err:
        print(f"Warning: Failed to load/process image for example {i} (ID: {item_id}): {img_err}. Skipping.")
        print(traceback.format_exc()) # More detail on image errors
        clear_memory()
        continue

    # --- Model Inference ---
    extracted_result = "Error" # Default value in case of error before extraction
    raw_output_text = "Error" # Store raw output separately
    try:
        # Prepare input for the VLM
        messages = [{"role": "user","content": [{"type": "image"},{"type": "text", "text": VLM_PROMPT}]}]
        # Use tokenizer for text part, processor combines text+image
        prompt_for_model = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
        inputs = processor(text=prompt_for_model, images=image_resized, return_tensors="pt").to(DEVICE)

        # Generate prediction
        print("Generating prediction...")
        with torch.no_grad(): # Essential for inference
            output_ids = model.generate(
                **inputs,
                max_new_tokens=20, # Keep low for single-letter answers
                do_sample=False,
                pad_token_id=tokenizer.eos_token_id # Important
            )

        # Decode the output
        input_len = inputs["input_ids"].shape[1]
        generated_ids = output_ids[:, input_len:]
        # Store the raw decoded output
        raw_output_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
        print(f"Raw model output: '{raw_output_text}'")

        # Extract the final answer using the updated function
        extracted_result = extract_predicted_answer(raw_output_text)
        print(f"Processed Answer (for JSONL): {extracted_result}") # Renamed variable for clarity

        # --- Store result ---
        # The key 'answer_key' will now hold either A-E or the raw text if extraction failed
        result_dict = {
            "id": item_id, # Key must be "id" for submission format
            "answer_key": extracted_result, # Use the result from extract_predicted_answer
            "language": language
            # Optional: Add raw output to JSONL for debugging if needed
            # "raw_output": raw_output_text
        }

        # --- Append result immediately to JSONL file ---
        append_result_to_file(OUTPUT_JSONL_FILENAME, result_dict)
        processed_count_this_run += 1
        processed_ids.add(item_id) # Add to set to avoid re-processing if script restarts

        # --- Explicit Memory Cleanup ---
        del inputs, output_ids, generated_ids, image_pil, image_resized, prompt_for_model, result_dict
        clear_memory()

    except Exception as e:
        print(f"âŒâŒâŒ Error processing example {i} (ID: {item_id}): {e}")
        print(traceback.format_exc()) # Print detailed traceback for errors
        # --- Cleanup even on error ---
        clear_memory()
        # Optionally, you could write an error placeholder to the file:
        # error_dict = {"id": item_id, "answer_key": f"ERROR: {e}", "language": language, "raw_output": raw_output_text}
        # append_result_to_file(OUTPUT_JSONL_FILENAME, error_dict)

    finally:
        item_duration = time.time() - start_time_item
        print(f"Finished example {i+1}. Time taken: {item_duration:.2f} seconds.")


# --- Final Summary and Conversion ---
end_time_loop = time.time()
total_duration = end_time_loop - start_time_loop

print("\n--- Prediction Loop Finished ---")
print(f"Total time: {total_duration:.2f} seconds")
print(f"Examples processed in this run: {processed_count_this_run}")
print(f"Examples skipped (already processed): {skipped_count}")
print(f"Total results expected in {OUTPUT_JSONL_FILENAME}: {len(processed_ids)}")

# 5. Convert JSONL to the final required JSON list format
print(f"\nConverting {OUTPUT_JSONL_FILENAME} to final {FINAL_JSON_FILENAME}...")
final_results_list = []
final_ids = set()
conversion_errors = 0
skipped_non_standard = 0 # Count entries skipped due to non A-E answer_key
try:
    with open(OUTPUT_JSONL_FILENAME, 'r', encoding='utf-8') as f_in:
        for line_num, line in enumerate(f_in):
            try:
                data = json.loads(line)
                # Basic validation of the loaded data from JSONL
                if 'id' in data and 'answer_key' in data and 'language' in data:
                    current_id = data['id']
                    answer_key = data['answer_key'] # This might be 'A', 'B', 'C', 'D', 'E' OR raw text

                    if current_id not in final_ids:
                         # --- Filter for final submission ---
                         # Only include entries where the answer_key is exactly A, B, C, D, or E
                         if answer_key in ["A", "B", "C", "D", "E"]:
                             final_results_list.append({
                                 "id": current_id,
                                 "answer_key": answer_key,
                                 "language": data['language']
                             })
                             final_ids.add(current_id)
                         else:
                              # This happens if extract_predicted_answer returned the raw text
                              print(f"Info: Skipping entry for ID {current_id} from final JSON because answer_key ('{answer_key}') is not A-E.")
                              skipped_non_standard += 1
                    else:
                         print(f"Warning: Duplicate ID '{current_id}' found on line {line_num+1}. Skipping.")
                         conversion_errors += 1
                else:
                    print(f"Warning: Skipping line {line_num+1} due to missing required keys (id, answer_key, language).")
                    conversion_errors += 1
            except json.JSONDecodeError:
                print(f"Warning: Skipping malformed JSON line {line_num+1} in {OUTPUT_JSONL_FILENAME}: {line.strip()}")
                conversion_errors += 1

    print(f"Loaded {len(final_results_list)} valid results (with A-E answers) for the final JSON file.")
    if skipped_non_standard > 0:
         print(f"Skipped {skipped_non_standard} entries from the final JSON because their 'answer_key' was not A-E (raw output was kept in JSONL).")
    if conversion_errors > 0:
        print(f"Encountered {conversion_errors} other errors/warnings during conversion (duplicates, missing keys, malformed lines).")

    # Check final size against total expected examples
    # Note: The final count might be *less* than total_examples if some answers couldn't be mapped to A-E
    print(f"Total examples in dataset: {total_examples}")
    print(f"Examples written to final JSON: {len(final_results_list)}")
    if len(final_results_list) + skipped_non_standard != len(processed_ids):
         print(f"WARNING: Mismatch in counts. Processed IDs: {len(processed_ids)}, Final JSON entries: {len(final_results_list)}, Skipped non-standard: {skipped_non_standard}. Check logs for conversion errors.")


    # Write the final JSON list
    with open(FINAL_JSON_FILENAME, "w", encoding="utf-8") as f_out:
        json.dump(final_results_list, f_out, indent=2, ensure_ascii=False) # Use indent=2 for readability
    print(f"Successfully created final submission file: {FINAL_JSON_FILENAME}")

except Exception as e:
    print(f"FATAL: Error during final conversion from JSONL to JSON: {e}")
    print(traceback.format_exc())

print("\n--- Script Finished ---")
```

```python
import json

# --- Configuration ---
input_jsonl_file = '/content/run_results_unstructured.jsonl'  # Name of your input JSONL file
output_json_file = '/content/run_results.json'  # Desired output JSON file name
# --- ---

all_data = []

print(f"Reading data from: {input_jsonl_file}")
try:
    # Open the input JSONL file for reading
    with open(input_jsonl_file, 'r', encoding='utf-8') as infile:
        for line_number, line in enumerate(infile, 1):
            # Remove leading/trailing whitespace (like newline characters)
            line = line.strip()
            if not line:  # Skip empty lines if any
                continue
            try:
                # Parse each line as a JSON object (which becomes a Python dict)
                data_object = json.loads(line)
                # Add the parsed dictionary to our list
                all_data.append(data_object)
            except json.JSONDecodeError as e:
                print(f"Warning: Skipping invalid JSON on line {line_number}: {e}")
                print(f"  Content: {line}")

except FileNotFoundError:
    print(f"Error: Input file '{input_jsonl_file}' not found.")
    exit(1) # Exit if input file is missing
except Exception as e:
    print(f"An unexpected error occurred while reading the input file: {e}")
    exit(1)

print(f"Writing {len(all_data)} records to: {output_json_file}")
try:
    # Open the output JSON file for writing
    with open(output_json_file, 'w', encoding='utf-8') as outfile:
        # Write the entire list of dictionaries as a single JSON array
        # indent=2 makes the output file readable (pretty-printed) like your example
        # ensure_ascii=False is important if your data contains non-ASCII characters
        json.dump(all_data, outfile, indent=2, ensure_ascii=False)
    print("Conversion successful!")

except IOError as e:
    print(f"Error: Could not write to output file '{output_json_file}': {e}")
    exit(1)
except Exception as e:
    print(f"An unexpected error occurred while writing the output file: {e}")
    exit(1)
```
