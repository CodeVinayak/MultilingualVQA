```python
import os
import json
from PIL import Image
from datasets import load_from_disk

# Load dataset
dataset_dir = "/content/dataset"  # Change this if needed
dataset = load_from_disk(dataset_dir)

# Define output folder and create it if necessary
output_folder = "/content/downloaded_images"
os.makedirs(output_folder, exist_ok=True)

# Maximum allowed pixels
MAX_PIXELS = 1280 * 28 * 28

# Initialize a dictionary to store image paths and their corresponding answer keys
results_dict = {}

# Process up to 200 images from the dataset
num_samples = min(200, len(dataset))
print(f"Processing {num_samples} images...")

for i, example in enumerate(dataset.select(list(range(num_samples)))):
    try:
        # Retrieve and convert the image to RGB
        img = example["image"]
        if not isinstance(img, Image.Image):
            print(f"Warning: Item {i} does not contain a valid image. Skipping.")
            continue
        img = img.convert("RGB")

        # Calculate new dimensions while maintaining aspect ratio
        width, height = img.size
        scale_factor = min((MAX_PIXELS / (width * height)) ** 0.5, 1.0)
        new_width = int(width * scale_factor)
        new_height = int(height * scale_factor)

        # Resize the image using LANCZOS resampling
        img_resized = img.resize((new_width, new_height), Image.Resampling.LANCZOS)

        # Define the new filename and save the resized image
        image_filename = f"image_{i}.png"
        image_path = os.path.join(output_folder, image_filename)
        img_resized.save(image_path, format="PNG", quality=100)

        # Save the mapping of image path to answer key
        answer_key = example["answer_key"]
        results_dict[image_path] = answer_key

        print(f"✅ Resized & saved: {image_filename} ({new_width}x{new_height}) | Answer: {answer_key}")

    except Exception as e:
        print(f"❌ Error processing image {i}: {e}")

# Save the results dictionary to a JSON file
results_json_path = os.path.join(output_folder, "results.json")
with open(results_json_path, "w") as json_file:
    json.dump(results_dict, json_file, indent=4)

print(f"\nResults JSON saved at: {results_json_path}")
```

```python
# =================================== app.py content starts here ===================================
with open("app.py", "w") as f:
    f.write("""
import streamlit as st
import torch
from PIL import Image
from transformers import AutoProcessor
from unsloth import FastVisionModel
from peft import PeftModel
import re
import traceback
import json # To load the dataset JSON
import os   # To check file existence
import random # To select a random image

# --- Configuration ---
DATASET_JSON_PATH = "/content/downloaded_images/results.json"

# --- Model Loading (Cached) ---
@st.cache_resource
def load_model_and_processor():
    st.write("Cache miss: Loading base model...")
    try:
        model, tokenizer = FastVisionModel.from_pretrained(
            "unsloth/Qwen2.5-VL-7B-Instruct-unsloth-bnb-4bit",
            load_in_4bit=True,
            use_gradient_checkpointing="unsloth"
        )
        st.write("Base model loaded. Loading adapter...")
        # Load adapter
        model = PeftModel.from_pretrained(model, "./vinayak_model")
        st.write("PEFT adapter loaded successfully.")

        # Load processor
        processor = AutoProcessor.from_pretrained("Qwen/Qwen2.5-VL-7B-Instruct")
        st.write("Processor loaded successfully.")
        return model, tokenizer, processor
    except Exception as e:
        st.error(f"Error during model/processor loading: {e}. Ensure adapter path './vinayak_model' is correct.")
        st.error(traceback.format_exc())
        st.stop()

# --- Dataset Loading (Cached) ---
@st.cache_data
def load_dataset_info(json_path):
    st.write(f"Cache miss: Loading dataset info from {json_path}...")
    if not os.path.exists(json_path):
        st.error(f"Dataset JSON file not found at: {json_path}")
        return None
    try:
        with open(json_path, 'r') as f:
            data = json.load(f)
        # Basic validation (check if it's a dictionary)
        if not isinstance(data, dict):
             st.error(f"Invalid format in {json_path}. Expected a JSON object (dictionary).")
             return None
        st.write(f"Loaded {len(data)} items from dataset JSON.")
        return data
    except json.JSONDecodeError as e:
        st.error(f"Error decoding JSON file {json_path}: {e}")
        return None
    except Exception as e:
        st.error(f"Error loading dataset file {json_path}: {e}")
        return None

# --- Explanation Generation ---
# Function to generate explanation for a KNOWN correct answer
def generate_explanation_for_correct_answer(model, tokenizer, processor, image, correct_answer_key):
    \"\"\"Generates an explanation focusing on why the pre-defined correct option is right.\"\"\"
    if not image or not correct_answer_key:
        return "Error: Image or correct answer key missing for explanation."
    if not all([model, tokenizer, processor]):
         return "Error: Model components not loaded."

    # Prompt asking specifically why the given answer is correct
    prompt_text = f\"\"\"Analyze the provided image carefully.

    Provide a detailed step-by-step reasoning explaining exactly why Option {correct_answer_key} is the correct answer based *only* on the visual evidence in the image. Focus solely on justifying Option {correct_answer_key}.
    \"\"\"
    messages = [{"role": "user", "content": [{"type": "image"}, {"type": "text", "text": prompt_text}]}]
    explanation_text = "Error generating explanation." # Default

    try:
        text_prompt = processor.apply_chat_template(messages, add_generation_prompt=True)
        inputs = processor(text=[text_prompt], images=[image], padding=True, return_tensors="pt").to("cuda" if torch.cuda.is_available() else "cpu")

        with torch.no_grad():
            output_ids = model.generate(
                **inputs,
                max_new_tokens=512, # Reduced tokens slightly for focused explanation
                do_sample=False,
                pad_token_id=processor.tokenizer.pad_token_id or processor.tokenizer.eos_token_id
            )

        full_decoded_output = processor.batch_decode(output_ids, skip_special_tokens=True)[0]
        gen_prompt_marker = "assistant\\n"
        marker_pos = full_decoded_output.rfind(gen_prompt_marker)
        if marker_pos != -1:
            explanation_text = full_decoded_output[marker_pos + len(gen_prompt_marker):].strip()
        else:
            explanation_text = full_decoded_output.strip() # Fallback

        # print("--- Explanation Generation Response (Debug) ---")
        # print(explanation_text)
        # print("--- End Explanation Generation Response ---")

    except Exception as e:
        explanation_text = f"Sorry, an error occurred during explanation generation: {e}"
        print(f"Error in generate_explanation_for_correct_answer: {e}")
        traceback.print_exc()

    return explanation_text

# Function for Upload mode: Model determines answer AND explains
def get_model_answer_and_explanation_upload(model, tokenizer, processor, image):
    \"\"\"Upload Mode: Asks model to determine answer (A/B/C/D) and explain its choice.\"\"\"
    if not image: return None, "Error: Image missing."
    if not all([model, tokenizer, processor]): return None, "Error: Model components not loaded."

    # Prompt asking model to decide and explain
    prompt_text = f\"\"\"Analyze the provided image carefully. Based on your analysis, determine which of the hypothetical options A, B, C, or D is the most accurate.

    First, state the option you determined is correct in the format 'Correct Answer: [X]' where X is A, B, C, or D.
    Second, provide a detailed step-by-step reasoning explaining *why* you chose that specific option based *only* on the visual evidence.
    \"\"\"
    messages = [{"role": "user", "content": [{"type": "image"}, {"type": "text", "text": prompt_text}]}]
    model_response_text = ""
    model_choice = None
    explanation = "Error during processing."

    try:
        text_prompt = processor.apply_chat_template(messages, add_generation_prompt=True)
        inputs = processor(text=[text_prompt], images=[image], padding=True, return_tensors="pt").to("cuda" if torch.cuda.is_available() else "cpu")

        with torch.no_grad():
            output_ids = model.generate(
                 **inputs, max_new_tokens=1024, do_sample=False,
                 pad_token_id=processor.tokenizer.pad_token_id or processor.tokenizer.eos_token_id
            )

        full_decoded_output = processor.batch_decode(output_ids, skip_special_tokens=True)[0]
        gen_prompt_marker = "assistant\\n"
        marker_pos = full_decoded_output.rfind(gen_prompt_marker)
        model_response_text = full_decoded_output[marker_pos + len(gen_prompt_marker):].strip() if marker_pos != -1 else full_decoded_output.strip()

        # print("--- Raw Model Response Upload Mode (Debug) ---")
        # print(model_response_text)
        # print("--- End Raw Model Response Upload Mode ---")

        match = re.search(r"Correct Answer:\s*([ABCD])", model_response_text, re.IGNORECASE)
        if match:
            model_choice = match.group(1).upper()
            explanation_start_index = match.end()
            explanation = model_response_text[explanation_start_index:].strip()
            explanation = re.sub(r"^[.:\\s]+", "", explanation) # Clean leading punctuation/space
            print(f"Parsed Model Choice (Upload): {model_choice}")
        else:
            print("Warning: Could not parse 'Correct Answer: [X]' pattern from model response (Upload).")
            explanation = f"Model response received, but could not determine the chosen option (A/B/C/D). Full response: {model_response_text}"

    except Exception as e:
        explanation = f"Sorry, an error occurred during model processing: {e}"
        print(f"Error in get_model_answer_and_explanation_upload: {e}")
        traceback.print_exc()

    return model_choice, explanation

# --- Helper Function to Load Image and Reset State ---
def load_image_and_reset(image_key):
    \"\"\"Loads image, gets answer, resets results state. Used by dataset nav/select.\"\"\"
    try:
        if not os.path.exists(image_key):
            st.error(f"Image file not found at path: {image_key}")
            st.session_state.current_image = None
            st.session_state.current_correct_answer = None
            st.session_state.selected_dataset_key = None
            st.session_state.current_dataset_index = -1 # Indicate invalid index
        else:
            st.session_state.current_image = Image.open(image_key).convert("RGB")
            st.session_state.selected_dataset_key = image_key
            st.session_state.current_correct_answer = st.session_state.dataset_data.get(image_key)
            st.session_state.current_dataset_index = st.session_state.dataset_keys.index(image_key)
            if not st.session_state.current_correct_answer:
                st.warning(f"Correct answer not found for {image_key} in JSON.")

        # Reset results for the new image
        st.session_state.user_choice = None
        st.session_state.show_result = False
        st.session_state.explanation = ""
        st.session_state.model_choice_upload = None

    except Exception as e:
        st.error(f"Error loading image {image_key}: {e}")
        st.session_state.current_image = None
        st.session_state.current_correct_answer = None
        st.session_state.selected_dataset_key = None
        st.session_state.current_dataset_index = -1

# --- Streamlit App UI and Logic ---
st.set_page_config(layout="wide")
st.title("Vision Language Model For Multilingual VQA")

# --- Load Model ---
# Ensure model is loaded once and stored in session state
if 'model_loaded' not in st.session_state:
    st.session_state.model, st.session_state.tokenizer, st.session_state.processor = load_model_and_processor()
    st.session_state.model_loaded = True
else:
    # Retrieve from session state if already loaded
    model = st.session_state.model
    tokenizer = st.session_state.tokenizer
    processor = st.session_state.processor

# --- Initialize State Variables ---
# Keep track of mode, current image, answers, etc.
if 'mode' not in st.session_state:
    st.session_state.mode = None # 'Dataset' or 'Upload'
if 'dataset_data' not in st.session_state:
    st.session_state.dataset_data = None
if 'dataset_keys' not in st.session_state:
    st.session_state.dataset_keys = []
if 'selected_dataset_key' not in st.session_state: # Stores the image path key
    st.session_state.selected_dataset_key = None
if 'current_dataset_index' not in st.session_state: # Stores the index in dataset_keys
    st.session_state.current_dataset_index = -1 # Initialize to -1 (no selection)
if 'current_image' not in st.session_state:
    st.session_state.current_image = None
if 'current_correct_answer' not in st.session_state: # Ground truth from JSON
    st.session_state.current_correct_answer = None
if 'user_choice' not in st.session_state:
    st.session_state.user_choice = None
if 'show_result' not in st.session_state:
    st.session_state.show_result = False
if 'explanation' not in st.session_state:
    st.session_state.explanation = ""
if 'model_choice_upload' not in st.session_state: # Model's choice in upload mode
    st.session_state.model_choice_upload = None
if 'upload_key_counter' not in st.session_state: # To reset file uploader
    st.session_state.upload_key_counter = 0


# --- Mode Selection ---
st.header("1. Select Mode")
mode_options = ["Select...", "Use Image from Dataset", "Upload Your Own Image"]
chosen_mode_display = st.radio(
    "Choose how to provide the image:",
    mode_options,
    key="mode_radio",
    horizontal=True,
    index=mode_options.index(st.session_state.mode) if st.session_state.mode in mode_options else 0
)

# Update state if mode changed, reset relevant parts
if chosen_mode_display != st.session_state.mode and chosen_mode_display != "Select...":
    st.session_state.mode = chosen_mode_display
    # Reset state when mode changes
    st.session_state.current_image = None
    st.session_state.selected_dataset_key = None
    st.session_state.current_correct_answer = None
    st.session_state.current_dataset_index = -1 # Reset index
    st.session_state.user_choice = None
    st.session_state.show_result = False
    st.session_state.explanation = ""
    st.session_state.model_choice_upload = None
    st.session_state.upload_key_counter += 1 # Increment key to force reset file uploader if switching away
    st.rerun() # Rerun to reflect mode change immediately


# --- UI for Dataset Mode ---
if st.session_state.mode == "Use Image from Dataset":
    st.header("2. Select Dataset Image")

    # Load dataset info if not already loaded
    if st.session_state.dataset_data is None:
        st.session_state.dataset_data = load_dataset_info(DATASET_JSON_PATH)
        if st.session_state.dataset_data:
            st.session_state.dataset_keys = list(st.session_state.dataset_data.keys())
            if not st.session_state.dataset_keys:
                 st.warning("Dataset JSON loaded, but it contains no image entries.")
                 st.session_state.dataset_data = None # Treat as empty
            else:
                 # --- Auto-load first image on mode entry ---
                 if st.session_state.current_dataset_index == -1: # Only if no image is selected yet
                     load_image_and_reset(st.session_state.dataset_keys[0])
                     st.rerun() # Rerun to display the first loaded image

    if not st.session_state.dataset_data or not st.session_state.dataset_keys:
        st.warning("Could not load dataset or dataset is empty. Please check the JSON file path and format.")
    else:
        num_images = len(st.session_state.dataset_keys)
        current_index = st.session_state.current_dataset_index
        current_key = st.session_state.selected_dataset_key

        # Display current image info
        if current_index != -1 and current_key:
            display_index = current_index + 1 # User-friendly 1-based index
            st.markdown(f"**Current Image:** {display_index} / {num_images} (`{os.path.basename(current_key)}`)")
        else:
            st.markdown("**Current Image:** None selected")

        # --- Navigation Buttons ---
        nav_cols = st.columns([1, 1, 1, 3]) # Adjust ratios as needed
        with nav_cols[0]:
            if st.button("⬅️ Previous", key="prev_img", disabled=(num_images <= 1 or current_index == -1), use_container_width=True):
                new_index = (current_index - 1 + num_images) % num_images # Wraps around
                load_image_and_reset(st.session_state.dataset_keys[new_index])
                st.rerun()
        with nav_cols[1]:
            if st.button("Next ➡️", key="next_img", disabled=(num_images <= 1 or current_index == -1), use_container_width=True):
                new_index = (current_index + 1) % num_images # Wraps around
                load_image_and_reset(st.session_state.dataset_keys[new_index])
                st.rerun()
        with nav_cols[2]:
            if st.button("🎲 Random", key="rand_img", disabled=(num_images <= 1), use_container_width=True):
                 # Ensure we pick a *different* random image if possible
                 possible_indices = list(range(num_images))
                 if num_images > 1 and current_index != -1:
                     possible_indices.pop(current_index)
                 new_index = random.choice(possible_indices)
                 load_image_and_reset(st.session_state.dataset_keys[new_index])
                 st.rerun()

        # --- Optional Dropdown for Direct Selection ---
        with nav_cols[3]:
             # Use a format function to show basename, easier to read
             def format_key(key):
                  idx = st.session_state.dataset_keys.index(key)
                  return f"{idx+1}. {os.path.basename(key)}"

             selected_key_from_dropdown = st.selectbox(
                 "Or jump to image:",
                 options=["Select..."] + st.session_state.dataset_keys,
                 key="dataset_selectbox",
                 index=st.session_state.current_dataset_index + 1 if st.session_state.current_dataset_index != -1 else 0, # +1 for "Select..."
                 format_func=lambda x: "Select..." if x == "Select..." else format_key(x)
             )

             # If a new image is selected from dropdown
             if selected_key_from_dropdown != "Select..." and selected_key_from_dropdown != st.session_state.selected_dataset_key:
                 load_image_and_reset(selected_key_from_dropdown)
                 st.rerun() # Rerun to display new image and reset buttons


# --- UI for Upload Mode ---
elif st.session_state.mode == "Upload Your Own Image":
    st.header("2. Upload Image")
    # Use the counter in the key to allow resetting
    uploaded_file = st.file_uploader(
        "Choose an image...",
        type=["jpg", "jpeg", "png"],
        key=f"file_uploader_{st.session_state.upload_key_counter}"
    )

    if uploaded_file is not None:
        try:
            new_image = Image.open(uploaded_file).convert("RGB")
            # Simple check: Assume new upload if filename changes or image wasn't set before
            is_new_upload = (st.session_state.current_image is None or
                             st.session_state.get("current_upload_name") != uploaded_file.name)

            if is_new_upload:
                print(f"New file uploaded: {uploaded_file.name}")
                st.session_state.current_image = new_image
                st.session_state.current_upload_name = uploaded_file.name
                st.session_state.current_correct_answer = None # No ground truth
                st.session_state.selected_dataset_key = None # Not from dataset
                st.session_state.current_dataset_index = -1 # Not from dataset

                # Reset results for new image
                st.session_state.user_choice = None
                st.session_state.show_result = False
                st.session_state.explanation = ""
                st.session_state.model_choice_upload = None
                st.rerun() # Rerun to display new image and clear old results

        except Exception as e:
            st.error(f"Error opening uploaded image: {e}")
            st.session_state.current_image = None
            # Also clear related state
            st.session_state.current_correct_answer = None
            st.session_state.selected_dataset_key = None
            st.session_state.current_dataset_index = -1


# --- Display Image and Options (If Image is Loaded) ---
if st.session_state.current_image:
    col1, col2 = st.columns([2, 3]) # Adjust column ratios if needed
    with col1:
        st.image(st.session_state.current_image, caption="Current Image", use_container_width=True)
        if st.session_state.mode == "Use Image from Dataset" and st.session_state.selected_dataset_key:
             st.caption(f"File: `{os.path.basename(st.session_state.selected_dataset_key)}`")
        elif st.session_state.mode == "Upload Your Own Image" and st.session_state.get("current_upload_name"):
             st.caption(f"File: `{st.session_state.current_upload_name}`")


    with col2:
        st.header("3. Choose Your Option")
        st.write("Select the option you think is correct:")

        option_keys = ["A", "B", "C", "D"]
        cols_buttons = st.columns(len(option_keys))

        button_clicked = False
        for i, key in enumerate(option_keys):
            with cols_buttons[i]:
                 # Disable buttons if results are already shown for the current image
                 if st.button(f"Choose {key}", key=f"btn_{key}", use_container_width=True, disabled=st.session_state.show_result):
                     st.session_state.user_choice = key
                     st.session_state.show_result = True
                     # Reset previous explanation/model choice before processing THIS choice
                     st.session_state.explanation = ""
                     st.session_state.model_choice_upload = None
                     button_clicked = True # Flag that a button press triggered this cycle

        # Rerun immediately after button click to process result and update UI (like disabling buttons)
        if button_clicked:
            st.rerun()

        # --- Process Answer and Display Feedback (only if a choice was made) ---
        if st.session_state.show_result and st.session_state.user_choice:
            st.markdown("---")
            st.subheader("4. Result & Explanation")

            user_choice = st.session_state.user_choice
            image = st.session_state.current_image

            # --- Dataset Mode Logic ---
            if st.session_state.mode == "Use Image from Dataset":
                correct_answer = st.session_state.current_correct_answer
                if not correct_answer:
                     # Should have been warned earlier, but handle defensively
                     st.error("Could not find the correct answer for this dataset image in the JSON file.")
                else:
                     is_correct = (user_choice == correct_answer)
                     if is_correct:
                         st.success(f"✅ Correct! You chose **{user_choice}**, which is the right answer according to the question.")
                         # Optionally, you could still generate explanation for the correct answer here
                         # if st.session_state.explanation == "":
                         #     # ... generate explanation ...
                         # st.info(f"**Assistant's Explanation (Why {correct_answer} is correct):**")
                         # st.write(st.session_state.explanation)
                     else:
                         st.error(f"❌ Incorrect. You chose **{user_choice}**, but the correct answer according to the question is **{correct_answer}**.")
                         # Generate explanation for the *dataset's* correct answer
                         if st.session_state.explanation == "": # Generate only once per incorrect choice display
                              with st.spinner(f"Generating explanation for why **{correct_answer}** is correct..."):
                                  explanation = generate_explanation_for_correct_answer(
                                      model, tokenizer, processor, image, correct_answer
                                  )
                                  st.session_state.explanation = explanation
                         st.info(f"**Assistant's Explanation (Why {correct_answer} is correct):**")
                         st.write(st.session_state.explanation)

            # --- Upload Mode Logic ---
            elif st.session_state.mode == "Upload Your Own Image":
                # Get model's answer and explanation only once per choice cycle
                if st.session_state.model_choice_upload is None and st.session_state.explanation == "":
                     with st.spinner("Assistant is analyzing the image and determining the best option..."):
                         model_choice, explanation = get_model_answer_and_explanation_upload(
                             model, tokenizer, processor, image
                         )
                         st.session_state.model_choice_upload = model_choice
                         st.session_state.explanation = explanation

                model_correct_answer = st.session_state.model_choice_upload
                explanation = st.session_state.explanation

                if model_correct_answer:
                    is_correct = (user_choice == model_correct_answer)
                    if is_correct:
                        st.success(f"✅ Correct! You chose **{user_choice}**, which matches the Assistant's conclusion.")
                    else:
                        st.error(f"❌ Incorrect. You chose **{user_choice}**, but the Assistant determined the correct option is **{model_correct_answer}**.")

                    # Always show model's explanation in upload mode after a choice is made
                    st.info(f"**Assistant's Explanation (Why it chose {model_correct_answer}):**")
                    st.write(explanation if explanation else "No explanation provided.")
                else:
                    # Case where model failed to specify A/B/C/D
                    st.warning("The Assistant analyzed the image but didn't specify a clear choice (A, B, C, or D).")
                    st.info("**Assistant's Full Response:**")
                    st.write(explanation if explanation else "No response received.")

            # --- Reset Button ---
            if st.button("Try Another Option / Reset Choice", key="reset_choice"):
                 st.session_state.user_choice = None
                 st.session_state.show_result = False
                 st.session_state.explanation = ""
                 # Keep model's choice for upload mode in case user tries another option on same image
                 # st.session_state.model_choice_upload = None # Don't reset this necessarily
                 st.rerun()

# Footer or Instructions
st.markdown("---")
st.caption("© Vinayak Kumar Singh 2025 | Fine-tuned for enhanced performance across diverse languages and image types.")
if st.session_state.mode == "Use Image from Dataset":
    st.caption(f"Dataset info loaded from: {DATASET_JSON_PATH}")
    if st.session_state.dataset_keys:
        st.caption(f"Use the navigation buttons (⬅️ ➡️ 🎲) or the dropdown to browse {len(st.session_state.dataset_keys)} images.")
elif st.session_state.mode == "Upload Your Own Image":
     st.caption("In Upload mode, the Model determines the 'correct' answer based on its analysis.")

""")
# =================================== app.py content ends here ===================================

print("'app.py' file created with Dataset navigation (Sequential/Random/Dropdown).")

# --- Execution Code (Keep this separate from the 'with open' block) ---
# (This part remains largely the same, ensuring Streamlit runs and ngrok connects)
import subprocess
import time
import os
import signal
import nest_asyncio
from pyngrok import ngrok, conf

print("Preparing to launch Streamlit and ngrok...")

# Kill existing processes if needed
try:
    result = subprocess.run(["fuser", "8501/tcp"], capture_output=True, text=True, check=False)
    pids = result.stdout.strip().split()
    if pids:
        print(f"Attempting to kill existing process(es) on port 8501: {pids}")
        for pid in pids:
            try:
                os.kill(int(pid), signal.SIGKILL)
                print(f"Killed PID {pid}.")
            except ProcessLookupError:
                print(f"PID {pid} not found.")
            except ValueError:
                 print(f"Invalid PID format: {pid}")
        time.sleep(2)
    else:
         print("No existing process found on port 8501.")
except FileNotFoundError:
    print("'fuser' command not found. Skipping process kill check.")
except Exception as e:
    print(f"An error occurred while trying to kill process on port 8501: {e}")


# Start Streamlit
print("Starting Streamlit app in the background...")
log_file = "streamlit_log.txt"
streamlit_proc = None
try:
    # Ensure the log file can be written to
    with open(log_file, 'w') as log_check:
         log_check.write("Streamlit log initialized.\n")

    streamlit_cmd = [
        "nohup", "streamlit", "run", "app.py",
        "--server.port=8501",
        "--server.headless=True",
        "--server.enableCORS=false",
        "--server.enableXsrfProtection=false",
    ]
    # Redirect stdout and stderr to the log file
    with open(log_file, 'a') as log: # Append to the initialized log
        streamlit_proc = subprocess.Popen(
            streamlit_cmd,
            stdout=log, stderr=subprocess.STDOUT,
            preexec_fn=os.setpgrp # Run in separate process group
        )
    print(f"Streamlit process started with PID: {streamlit_proc.pid if streamlit_proc else 'Error'}. Logs in '{log_file}'")
    print("Waiting for Streamlit to initialize (including model load)...")
    # Increased wait time as model loading now happens inside Streamlit via cache
    time.sleep(35) # Wait 35 seconds (adjust if model loading takes longer)
    # Basic check if process terminated early
    if streamlit_proc and streamlit_proc.poll() is not None:
         raise RuntimeError(f"Streamlit process terminated unexpectedly with code {streamlit_proc.poll()}. Check '{log_file}'.")
    print("Streamlit seems to be running. Attempting ngrok connection.")

except Exception as e:
    print(f"❌ Error starting Streamlit: {e}")
    if os.path.exists(log_file):
         print(f"--- Contents of {log_file} ---")
         try:
             with open(log_file, 'r') as log:
                  print(log.read())
         except Exception as log_e:
              print(f"Error reading log file: {log_e}")
         print("--- End Log ---")
    raise # Stop the cell


# Start ngrok
nest_asyncio.apply()
print("Setting up ngrok tunnel...")

# Disconnect existing tunnels
try:
    tunnels = ngrok.get_tunnels()
    for tunnel in tunnels:
        print(f"Disconnecting existing ngrok tunnel: {tunnel.public_url}")
        ngrok.disconnect(tunnel.public_url)
    time.sleep(1)
except Exception as e:
    print(f"Info: Error disconnecting ngrok tunnels (may be none active): {e}")

# Connect ngrok
try:
    # Optional: Add auth token here if needed
    # conf.get_default().auth_token = "YOUR_NGROK_AUTH_TOKEN"
    tunnel = ngrok.connect(addr="8501", proto="http")
    url = tunnel.public_url
    print("-----------------------------------------------------")
    print(f"✅ Streamlit App should be live at: {url}")
    print("-----------------------------------------------------")


except Exception as e:
    print(f"❌ Error starting ngrok tunnel: {e}")
    print("Check ngrok installation, authentication, and if Streamlit is running correctly on port 8501.")
    print(f"Review '{log_file}' for potential Streamlit errors preventing connection.")


# Keep cell running
# (No explicit loop needed, cell just needs to stay active)
```
